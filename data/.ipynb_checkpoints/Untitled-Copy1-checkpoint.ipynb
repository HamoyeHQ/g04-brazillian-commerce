```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from googletrans import Translator

%matplotlib inline
```

```python
#identifying the datasets in local storage

dataset_files = glob.glob('*.csv')
dataset_files
```

```python
datasets_dict = {}
```

```python
#loading the dataset files into pandas dataframes

for file in dataset_files:
    dataset_name = file.strip('dataset.csv')+'df'
    df = pd.read_csv(file)
    datasets_dict[dataset_name] = df
```

```python
datasets_dict['olist_customers_df']
```

```python
customers_df = datasets_dict['olist_customers_df']
geolocation_df = datasets_dict['olist_geolocation_df']
orders_df = datasets_dict['olist_orders_df']
order_items_df = datasets_dict['olist_order_items_df']
order_payments_df = datasets_dict['olist_order_payments_df']
order_reviews_df = datasets_dict['olist_order_reviews_df']
products_df = datasets_dict['olist_products_df']
sellers_df = datasets_dict['olist_sellers_df']
product_category_name_translation_df = datasets_dict['product_category_name_translationdf']
```

```python
customers_order_df = pd.merge(customers_df, orders_df, on='customer_id', how='left')
customers_order_df
```

```python
for col in ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']:
    customers_order_df[col] = pd.to_datetime(customers_order_df[col])
```

```python
customers_order_items_df = pd.merge(customers_order_df, order_items_df, on='order_id', how='left')
customers_order_items_df
```

```python
customers_order_items_df = pd.merge(customers_order_items_df, products_df, on='product_id', how='left')
customers_order_items_df
```

```python
customers_order_items_df = pd.merge(customers_order_items_df, product_category_name_translation_df, on='product_category_name', how='left')
customers_order_items_df
```

```python
customers_order_items_df = pd.merge(customers_order_items_df, sellers_df, on='seller_id', how='left')
customers_order_items_df
```

```python
customers_order_items_review_df = pd.merge(customers_order_items_df, order_reviews_df, on='order_id', how='left')
customers_order_items_review_df
```

```python
customers_order_payment_df = pd.merge(customers_order_items_df, order_payments_df, on='order_id', how='left')
customers_order_payment_df.head()
```

```python
product_items_df=customers_order_items_review_df.drop(columns=[
                                                         'shipping_limit_date',
                                                         'product_category_name',
                                                         'product_name_lenght',
                                                         'product_description_lenght',
                                                         'product_photos_qty',
                                                         'review_score',
                                                         'review_id',
                                                         'review_comment_title', 
                                                         'review_comment_message',
                                                         'review_creation_date',
                                                         'review_answer_timestamp'])
product_items_df.head()
```

```python
product_items_df.columns
```

```python
#checking if weight affects delivery cost, checking delivery within the same state
order_within_states_df = pd.DataFrame(product_items_df[product_items_df['customer_state'] == product_items_df['seller_state']])
order_within_states_df
```

```python
#checking if the price affects delivery cost
order_within_states_price_df = order_within_states_df.groupby(['order_id','product_category_name_english','price']).freight_value.mean().sort_values(ascending=False)
order_within_states_price_df
```

```python
#checking if the weight affects delivery cost
order_within_states_weight_df = order_within_states_df.groupby(['order_id','product_category_name_english','product_weight_g']).freight_value.mean().sort_values(ascending=False)
order_within_states_weight_df
```
